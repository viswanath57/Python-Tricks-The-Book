```python
class Repeater:
    def __init__(self, value):
        self.value = value
    def __iter__(self):
        return RepeatIterator(self)

```

```python
class RepeatIterator:
    def __init__(self, source):
        self.source = source
    def __next__(self):
        return self.source.value
    
```

```python
repeater = Repeater("Hello")
for item in repeater:
    print(item)
```
